```python
df_country= pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
#Combine
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

**# Maximum number of transcation probably will be in India**

```python
country_names= final_df.Country.value_counts().index
country_values=final_df.Country.value_counts().values
```

```python
#Pie Chart
plt.pie(x=country_values,labels=country_names)
```

```python
#Plotting the top 3 country that uses Zomato
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.2f%%')
```

**Observation**-`Most of the transcation of zomato are from Indian Restaurant, after that United states and then United Kingdom`

```python
#Numerical Varibles
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

**Observation**-

1. When rating is between 4.5 to 4.9 --->Excellent
2. When the ratings are between 4.0-4.4 ---> Very Good
3. if the rating is between 3.5-3.9 then ----> Good
4. if the rating is between 3.0-3.4 then Average
5. Anything below 2.5 is poor







```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings)
```

```python
# Color has been not mapped
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

**Observation** -
1. Not rated has highest count
2. max no of rating are between 2.5 to 3.4
