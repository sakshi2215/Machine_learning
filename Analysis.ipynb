```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
df = pd.read_csv('/content/Churn_Modelling.csv')
df.head(5)
```

```python
df_new = df[['Geography','Gender','Exited']]
df_new.head(5)
```

**Seaborn Countplot**

```python
df_new.Exited.value_counts()
```

```python
for i , predictor in enumerate(df_new.drop(columns=['Exited'])):
  plt.figure()
  sns.countplot(data=df_new,x=predictor,hue='Exited')
```

Female and germany are more Churners.

**Histplot**

```python
sns.histplot(x='Gender',hue='Geography',data= df_new,stat='count',multiple='dodge')
```

```python
#Xhurned Customer
new_df1=df_new.loc[df_new['Exited']==1]
```

```python
sns.histplot(x='Gender',hue='Geography',data= new_df1,stat='count',multiple='dodge')
```

Female Germens are more churners than male. Female-36% and Male- 29%

```python
tot=sns.kdeplot(df.Age[(df['Exited']==0)],
            color='Red',shade=True)
tot=sns.kdeplot(df.Age[(df['Exited']==1)],
            color='Blue',shade=True)
tot.legend(['No Churn','Churned'],loc='upper right')
tot.set_ylabel('Density')
tot.set_xlabel('Age')
tot.set_title('Age by churn')
```
